```python
# imports
import pandas as pd
import yfinance as yf
from datetime import date, timedelta
from csv import reader
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# dates, today and one year ago
today =  date.today()
todaystring = str(today)
year_ago =  today - timedelta(days=365)
yesterday =  today - timedelta(days=1)
two_days_ago =  today - timedelta(days=2)
```

```python
# import the stocks I want to use
open_file = open('Stock Market Data.csv', encoding='utf8')
read_file = reader(open_file)
Stocks = list(read_file)
stock_header = Stocks[0]
stocks = Stocks[1:]
```

```python
# creating a ticker list
ticker_list = []
for row in stocks:
    ticker_list.append(row[1])
    
print(ticker_list)
```

```python
# function that returns list containing stock highs from last 365 days
def stock_period_data(tckr, startDate):
    tckr = str(tckr)
    data = yf.download(tckr, startDate)
    return [data['High'], data['Low'], data['Adj Close']]
```

```python
# Plot the data for the stock
def plotStock(plotdata, tckr, high, low, startDate):
    fib786 = high - ((high - low)*0.7860)
    fib618 = high - ((high - low)*0.6180)
    fib382 = high - ((high - low)*0.3820)
    fib236 = high - ((high - low)*0.2360)
    
    x_coord = [startDate, today]
    y_coord_fib786 = [fib786, fib786]
    y_coord_fib618 = [fib618, fib618]
    y_coord_fib382 = [fib382, fib382]
    y_coord_fib236 = [fib236, fib236]
    
    #startDate = str(startDate)
    tckr = str(tckr)
    #data = yf.download(tckr, startDate)
    #data['Adj Close'].plot(label = tckr)
    plotdata.plot(label = tckr, figsize = (12,5))
    
    plt.plot(x_coord, y_coord_fib236, label = '23.6%')
    plt.plot(x_coord, y_coord_fib382, label = '38.2%')
    plt.plot(x_coord, y_coord_fib618, label = '61.8%')
    plt.plot(x_coord, y_coord_fib786, label = '78.6%')
    
    plt.legend(bbox_to_anchor = (1.1,.5))
    plt.show()
```

```python
# The goal in the following cells is to create a dictionary containing stocks' and their 52 week highs

```

```python
# finds high price over time period
def find_high(stock):
    high = 0
    for row in stock:
        if high < row:
            high = row
    return high

# finds low price over time period 
def find_low(stock):
    low = 10000
    for row in stock:
        if low >= row:
            low = row
    return low
```

```python
#ploting stocks as well as creating dictionaries for 52 week highs and lows
# create dictionary with tickers and their 52 week high
high_dict = {}

# create dictionary with tickers and their 52 week low
low_dict = {}

for row in ticker_list:
    
    period_data = stock_period_data(row, year_ago);
    
    stock_high = find_high(period_data[0])
    high_dict[row] = stock_high
    
    stock_low = find_low(period_data[1])
    low_dict[row] = stock_low
    
    plotStock(period_data[2], row, stock_high, stock_low, year_ago)
```

```python
print('Highs: \n')
print(high_dict)
print('\nLows: \n')
print(low_dict)
```

```python
penny = "OGI"
penny_period_data = stock_period_data(penny, year_ago)
penny_high = find_high(penny_period_data[0])
penny_low = find_low(penny_period_data[1])
print(penny_high)
print(penny_low)
plotStock(penny_period_data[2], penny, penny_high, penny_low, year_ago)
```
